```python
orders['Order_Date']=pd.to_datetime(orders['Order_Date'])
```

```python
orders.info()
```

```python
orders['Month']=orders['Order_Date'].dt.strftime("%B")  #%A for day n %B for month
orders
```

```python
plt.figure(figsize=(10,3))
sns.countplot(x='Month',data=orders,hue="Month",palette="viridis")
plt.xticks(rotation=30)
plt.show()
```

```python
orders['Hour']=orders["Order_Date"].dt.strftime("%H")
orders.info()
```

```python
orders["Hour"]=orders["Hour"].astype("int")
orders.info()
```

```python
sns.lineplot(x="Hour",y="Total_Amount",data=orders,estimator="sum") #wihtout estimator as sum it gives avg,but for total value we use estimator as sum
plt.show()
```

```python
print(orders)
print(customers)
```

```python
df=pd.merge(left=orders,right=customers,on="Customer_ID",how="inner")
df
```

```python
plt.figure(figsize = (12,5)) #width, height
sns.countplot(x = "Customer_Location", data = df, hue = "Customer_Age_Group", palette = "viridis", edgecolor = "black")
plt.legend(bbox_to_anchor = (1,1))
plt.show()
```
